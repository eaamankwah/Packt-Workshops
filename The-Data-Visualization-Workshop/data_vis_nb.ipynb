```python
import matplotlib.pyplot as plt 
import numpy as np 

np.random.seed(19) 
center = np.ones(25) * 50 
spread = np.random.rand(50) * 100 

outlier_up = np.random.rand(10) * 100 + 100 
outlier_down = np.random.rand(10) * - 100 
data = np.concatenate((spread, center, outlier_up, outlier_down)) 

fig, ax = plt.subplots() 
ax.set_title('Box Plot') 
ax.boxplot(data) 

option_dict = dict(markerfacecolor='b', marker='D')
ax.boxplot(data, vertical=True, flierprops=option_dict)

plt.show() 
```
